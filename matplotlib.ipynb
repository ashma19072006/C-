#1)Line plot

```python
import matplotlib.pyplot as plt
x=['mon','tues','wed','thurs','fri','sat','sun']
y=[10,5,30,20,6,19,26]
plt.plot(x,y,c='g',marker='o')
plt.title('Bakery Sales This Week')
plt.xlabel('days')
plt.ylabel('sales')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['mon','tues','wed','thurs','fri','sat','sun']
y=[10,5,30,20,6,19,26]
plt.plot(x,y,c='g',marker='o')
plt.title('Bakery Sales This Week')
plt.xlabel('days')
plt.ylabel('sales')
plt.grid(axis='y',ls='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['mon','tues','wed','thurs','fri','sat','sun']
y=[10,5,30,20,6,19,26]
plt.plot(x,y,ls='dotted',marker='s',c='m',linewidth='3')

font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('Bakery Sales This Week',fontdict=font1)
plt.xlabel('days',fontdict=font2)
plt.ylabel('sales',fontdict=font2)
plt.grid()
plt.show()
#ls=linestyle
# Style       	          Or
# 'solid' (default)	     '-'
# 'dotted'	             ':'
# 'dashed'	             '--'
# 'dashdot'	             '-.'
# 'None'	             '' or ' '

```

```python
import matplotlib.pyplot as plt

days = ['mon', 'tues', 'wed', 'thurs', 'fri', 'sat', 'sun']

sales_bakery = [10, 5, 30, 20, 6, 19, 26]
sales_cafe   = [8, 12, 25, 18, 10, 22, 30]
sales_stalls =[5,3,8,9,10,25,28]
plt.plot(days, sales_bakery, marker='o', label='Bakery Sales')
plt.plot(days, sales_cafe, marker='s', label='Cafe Sales')
plt.plot(days,sales_stalls,marker='*',label='stall sales')

plt.title('Weekly Sales Comparison')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend()   # ye lines ko identify karne ke liye
# plt.grid(True)

plt.show()
```

#2)Scatter Plot

#3)Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([25,40,15,20])


plt.pie(x,startangle=90)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([25,40,15,20])
y=np.array(['apple','banana','grapes','dates'])
my =[0.2,0,0,0]
plt.pie(x,labels=y,startangle=90,explode=my,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([25,40,15,20])
y=np.array(['apple','banana','grapes','dates'])
plt.pie(x,labels=y,startangle =90,autopct='%1.0f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([25,20,15,40])
y=np.array(['apple','banana','grapes','dates'])

plt.pie(x,labels=y,startangle =90,autopct='%1.0f%%',shadow=True)
plt.legend(title ="four fruits:")
plt.show()
```

#4)Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([10,20,14,30,25,40])
x=np.array(['A','B','C','D','E','F'])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([10,20,14,30,25,40])
x=np.array(['A','B','C','D','E','F'])
plt.barh(x,y,color='green')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D','E','F'])
x_pos = np.arange(len(x))   # numeric positions

day1 = np.array([10,20,14,30,25,40])
day2 = np.array([15,26,18,25,15,30])

width = 0.35

plt.bar(x_pos - width/2, day1, width, label='Day-1')
plt.bar(x_pos + width/2, day2, width, label='Day-2')

plt.xticks(x_pos, x)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Comparison Between Day-1 and Day-2')
plt.legend()

plt.show()
```
