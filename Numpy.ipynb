indexing and slicing


```python
#indexing and slicing for 1-d array

#indexing
import numpy as np
x =np.array([1,2,3,4,5])
print(x[1])
print(x[-1])#last element

#slicing:-extracting subset of data
#arr[start:stop:step]
print(x[0:2:1])
print(x[::-1])
```

```python
#indexing and slicing for 2-d array

#indexing
import numpy as np
x =np.array([[1,2,3],[2,3,4]])
print(x[1][2])
print(x[-1][-1])#last element

#slicing:-extracting subset of data
print(x[0:1])
```

```python
#fancy indexing
import numpy as np
x = np.array([10,20,30,40,50,60])
print(x[[0,2,4]])


#filtering data:-boolean masking
print(x[x>25])
```

##Handling misssing values

i)np.isnan(array):-detect missing values . nan(not a number)

```python
import numpy as np

arr =np.array([1,2,np.nan,4,np.nan])
print(np.isnan(arr))

##we can't compare nan values directly
##print(np.nan == np.nan)  wrong
```

ii)np.nan_to_num(array,nan =value) default =0

```python
import numpy as np

arr =np.array([1,2,np.nan,4,np.nan])
cleaned_arr =np.nan_to_num(arr,nan=1)
print(cleaned_arr)
```

iii)np.isinf(array):-

```python
import numpy  as np

arr =np.array([1,2,np.inf,4,-np.inf])
print(np.isinf(arr))
```

```python
import numpy  as np

arr =np.array([1,2,np.inf,4,-np.inf])
print(np.isinf(arr))

cleaned_arr =np.nan_to_num(arr,posinf=100,neginf=-100)
print(cleaned_arr)
```
